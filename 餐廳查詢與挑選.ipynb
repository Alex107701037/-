```python
import csv
import random
with open('餐廳資料.csv', encoding='utf-8',newline='') as csvfile:
    df=csv.reader(csvfile)
    all_res=[]
    res_info=[]
    a,b,c,d,e,f=[],[],[],[],[],[]
    a1,b1,c1,d1,e1,f1=[],[],[],[],[],[]
    for row in df:
        a.append(row[0])
        b.append(row[1])
        c.append(row[2])
        d.append(row[3])
        e.append(row[4])
        f.append(row[5])
        all_res.append(row)
    csvfile.close()


while True:
    all_res_name=[]
    for i in all_res:
        all_res_name.append(i[1])
    function=input('查詢店名請輸入1\n利用分類查詢餐廳請輸入2\n新增餐廳請輸入3\n更改餐廳請輸入4\n刪除餐廳請輸入5\n隨機挑選餐廳請輸入6\n結束請輸入0\n')
    if function=='1':
        search_name=input('請輸入店名:')
        if search_name in all_res_name:
            for i in all_res:
                if search_name ==i[1]:
                    print(i[0]+' '+i[1]+' '+i[2]+' '+i[3]+' '+i[4])
        else:
            print('沒有此店名。')
        
    elif function=='2':
        search=input('請輸入風格：')
        if search in ['義式','美式','台式','東南亞','日式' , '韓式']:
            for i in all_res:
                if search==i[0]:
                    print(i[0]+' '+i[1]+' '+i[2]+' '+i[3]+' '+i[4])
        else:
            print('輸入錯誤，請重新輸入。')
    elif function=='3':
        new_one=input('請輸入風格：')
        if new_one in ['義式','美式','台式','東南亞','日式' , '韓式']:
            new_one1=input('請新增店名：')
            new_one2=input('請新增營業時間：')
            new_one3=input('請新增工作天：')
            new_one4=input('請新增電話：')
            new_one5=input('請新增網址：')
            all_res=all_res+[[new_one,new_one1,new_one2,new_one3,new_one4,new_one5]]
            print('新增完成。')
        else:
            print('此風格不在清單內。')
    elif function=='4':
        edit=input('請輸入風格：')
        if edit in ['義式','美式','台式','東南亞','日式' , '韓式']:
            edit1=input('請輸入店名：')
            if edit1 in all_res_name:
                edit2=input('請更改營業時間：')
                edit3=input('請更改工作天：')
                edit4=input('請更改電話：')
                edit5=input('請更改網址：')
                for i in all_res:
                    if edit1==i[1]:
                        i[0]=i[0]
                        i[1]=i[1]
                        i[2]=edit2
                        i[3]=edit3
                        i[4]=edit4
                        i[5]=edit5
                print('更改成功。')
            else:
                print('沒有此店名，請重新輸入。')
        
        else:
            print('沒有此風格，請重新輸入')
    elif function=='5':
        delete=input('請輸入店名：')
        if delete in all_res_name:
            for i in all_res:
                if delete==i[1]:
                    all_res.remove(i)
            print('刪除完成')
        
        else:
            print('沒有此店名，請重新輸入。')
    elif function=='6':
        random_choice=random.sample(all_res,3)
        for i in random_choice:
            print(i[0]+' '+i[1]+' '+i[2]+' '+i[3]+' '+i[4])
    elif function=='0':
        print('結束程式')
        break

    else:
        print('輸入錯誤，請重新輸入。')




newcsv=open('餐廳資料.csv','w',encoding='utf-8')
csvwriter = csv.writer(newcsv,dialect = ("excel"))
for i in all_res:
    csvwriter.writerow(i)
    
newcsv.close()
```
